* for the highest traffic stations

#### Importing packages 

```python
import pandas as pd
import numpy as np
import pickle

import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_formats = ['svg']  # or retina
%matplotlib inline

sns.set(context='notebook', 
    style='whitegrid', 
    font_scale=1.1)
```

#### Loading pickled dataframes 

```python
day_of_week_avg = pd.read_pickle('day_of_week_avg.pickle')
day_of_week_avg2 = pd.read_pickle('day_of_week_avg2.pickle')
```

#### Graph

```python
plt.figure(figsize=(10,5))

plt.scatter(day_of_week_avg['DAY_OF_WEEK_NUM'], day_of_week_avg['DAILY_ENTRIES_EXITS'])
plt.scatter(day_of_week_avg2['DAY_OF_WEEK_NUM'], day_of_week_avg2['DAILY_ENTRIES_EXITS'])
plt.xlabel('Day of the Week')
plt.ylabel('Average Daily Foot Traffic')
plt.title('Comparing Foot Traffic for Days of the Week 2019-2020\n(For the Highest Traffic Stations)', weight='bold', fontsize=15, y=1.1);

plt.xticks(np.arange(7),['Mo','Tu','We','Th','Fr','St','Sn'])
plt.yticks(np.linspace(0,250000, 6), ['0','50000', '100000','150000','200000','250000'])
           
plt.legend(['Jan - March 2019 Data','Jan - March 2021 Data'],loc = 0);
```

I plotted two scatterplots, one with data from Jan-March 2019 and one with data from Jan-March 2021, to compare the average daily station traffic before and during Covid (for the top 10 highest stations).

I am hoping to turn this graph into a bar chart, where the 2019 and 2021 data is displayed side by side, but this graph still shows that traffic has significantly declined in 2021 (during Covid), particularly during the working week. 

My target client (the liveboard advertising company, OutFront Media) may be interested to know that, while there is usually a large difference in station traffic between the weekdays and the weekends, during Covid, there is not. Therefore they may not be able to charge a premium price for liveboard space during the working week during Covid.
